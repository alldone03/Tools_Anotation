```python
import cv2

# Fungsi scale
def scale_image(img, scale_x, scale_y):
    width = int(img.shape[1] * scale_x)
    height = int(img.shape[0] * scale_y)
    return cv2.resize(img, (width, height))

# Fungsi crop relatif
def crop_relative(img, x_rel, y_rel, w_rel, h_rel):
    h, w = img.shape[:2]
    x = int(w * x_rel)
    y = int(h * y_rel)
    width = int(w * w_rel)
    height = int(h * h_rel)
    return img[y:y+height, x:x+width]

# Baca gambar asli
image_original = cv2.imread('C:/Users/Aldan/Desktop/IMPROTOYOTA/Camera 1/Zen/AR/CAMERA_1_2024-08-27_08-41-52.jpg')  

# Callback saat trackbar digerakkan
def update(val=0):
    scale_flag = cv2.getTrackbarPos('Scale On/Off', 'Crop')
    x_rel = cv2.getTrackbarPos('x', 'Crop') / 100
    y_rel = cv2.getTrackbarPos('y', 'Crop') / 100
    w_rel = cv2.getTrackbarPos('width', 'Crop') / 100
    h_rel = cv2.getTrackbarPos('height', 'Crop') / 100

    # Gunakan gambar yang di-scale atau tidak
    if scale_flag == 1:
        image = scale_image(image_original, 1.5, 1.5)
    else:
        image = image_original.copy()

    cropped = crop_relative(image, x_rel, y_rel, w_rel, h_rel)

    # Buat preview dengan kotak crop
    preview = image.copy()
    h_img, w_img = preview.shape[:2]
    x = int(x_rel * w_img)
    y = int(y_rel * h_img)
    w = int(w_rel * w_img)
    h = int(h_rel * h_img)
    cv2.rectangle(preview, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Tampilkan hasil
    cv2.imshow('Crop', preview)
    cv2.imshow('Hasil Crop', cropped)

# Buat jendela dan slider
cv2.namedWindow('Crop')
cv2.createTrackbar('Scale On/Off', 'Crop', 0, 1, update)
cv2.createTrackbar('x', 'Crop', 29, 100, update)
cv2.createTrackbar('y', 'Crop', 29, 100, update)
cv2.createTrackbar('width', 'Crop', 39, 100, update)
cv2.createTrackbar('height', 'Crop', 28, 100, update)

# Panggil pertama kali
update()

# Loop sampai ESC ditekan
while True:
    if cv2.waitKey(1) == 27:
        break

cv2.destroyAllWindows()
```

```python
def adjust_label_to_crop(label, img_shape, x_rel, y_rel, w_rel, h_rel):
    img_h, img_w = img_shape[:2]

    # Koordinat crop dalam piksel
    crop_x = int(img_w * x_rel)
    crop_y = int(img_h * y_rel)
    crop_w = int(img_w * w_rel)
    crop_h = int(img_h * h_rel)

    x, y, w, h = label  # label dalam piksel

    # Hitung koordinat baru relatif terhadap crop
    new_x = x - crop_x
    new_y = y - crop_y

    # Periksa apakah label berada di dalam area crop
    if new_x < 0 or new_y < 0 or new_x + w > crop_w or new_y + h > crop_h:
        return None  # Label di luar crop, bisa di-skip atau dipotong

    return (new_x, new_y, w, h)
```

```python
import cv2
import os

# --- Konfigurasi ---
input_folder = "C:/Users/Aldan/Desktop/ImproTYT/dataset"
output_folder = "C:/Users/Aldan/Desktop/ImproTYT/dataset_cropped"
scale_x, scale_y = 0, 0
# scale_x, scale_y = 1.5, 1.5
# crop_rel = (0.27, 0.20, 0.43, 0.47)  # x_rel, y_rel, w_rel, h_rel
crop_rel = (0.25, 0.21, 0.46, 0.46)  # x_rel, y_rel, w_rel, h_rel
# crop_rel = (0.30, 0.27, 0.41, 0.41)  # x_rel, y_rel, w_rel, h_rel

os.makedirs(output_folder, exist_ok=True)

# --- Fungsi ---
def scale_image(img, scale_x, scale_y):
    w = int(img.shape[1] * scale_x)
    h = int(img.shape[0] * scale_y)
    return cv2.resize(img, (w, h))

def crop_relative(img, x_rel, y_rel, w_rel, h_rel):
    h, w = img.shape[:2]
    x = int(w * x_rel)
    y = int(h * y_rel)
    cw = int(w * w_rel)
    ch = int(h * h_rel)
    return img[y:y+ch, x:x+cw], x, y, cw, ch

def adjust_label(label_line, img_w, img_h, crop_x, crop_y, crop_w, crop_h):
    cls, x, y, w, h = map(float, label_line.strip().split())
    abs_x = x * img_w
    abs_y = y * img_h
    abs_w = w * img_w
    abs_h = h * img_h

    # Hitung posisi baru
    new_x = abs_x - crop_x
    new_y = abs_y - crop_y

    if new_x < 0 or new_y < 0 or new_x > crop_w or new_y > crop_h:
        return None  # Label keluar dari crop

    # Normalisasi ulang ke ukuran crop
    norm_x = new_x / crop_w
    norm_y = new_y / crop_h
    norm_w = abs_w / crop_w
    norm_h = abs_h / crop_h

    return f"{int(cls)} {norm_x:.6f} {norm_y:.6f} {norm_w:.6f} {norm_h:.6f}"

# --- Proses semua file ---
for file in os.listdir(input_folder):
    if file.endswith('.jpg') or file.endswith('.png'):
        img_path = os.path.join(input_folder, file)
        label_path = os.path.splitext(img_path)[0] + '.txt'

        image = cv2.imread(img_path)
        if scale_x == 0 or scale_y == 0:
            scaled = image.copy()
        else:
            scaled = scale_image(image, scale_x, scale_y)
        cropped, cx, cy, cw, ch = crop_relative(scaled, *crop_rel)

        # Simpan gambar hasil crop
        out_img_path = os.path.join(output_folder, file)
        cv2.imwrite(out_img_path, cropped)

        # Proses label jika ada
        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                lines = f.readlines()

            new_labels = []
            for line in lines:
                new_line = adjust_label(line, scaled.shape[1], scaled.shape[0], cx, cy, cw, ch)
                if new_line:
                    new_labels.append(new_line)

            # Simpan label baru
            out_label_path = os.path.splitext(out_img_path)[0] + '.txt'
            with open(out_label_path, 'w') as f:
                f.write('\n'.join(new_labels))
```

```python
import os
import shutil
import yaml

def read_labels_from_yaml(yaml_path):
    with open(yaml_path, 'r') as f:
        data = yaml.safe_load(f)
        return data.get('names', [])

def merge_roboflow_dataset(source_folder, output_folder):
    yaml_path = os.path.join(source_folder, "data.yaml")
    train_images = os.path.join(source_folder, "train", "images")
    train_labels = os.path.join(source_folder, "train", "labels")

    os.makedirs(output_folder, exist_ok=True)

    # Buat labels.txt dari data.yaml
    if os.path.isfile(yaml_path):
        class_names = read_labels_from_yaml(yaml_path)
        with open(os.path.join(output_folder, "labels.txt"), "w") as f:
            for name in class_names:
                f.write(name + "\n")
        print("labels.txt berhasil dibuat dari data.yaml")

    # Salin semua gambar
    if os.path.isdir(train_images):
        for file in os.listdir(train_images):
            if file.endswith((".jpg", ".jpeg", ".png")):
                shutil.copy(os.path.join(train_images, file), output_folder)

    # Salin semua label .txt
    if os.path.isdir(train_labels):
        for file in os.listdir(train_labels):
            if file.endswith(".txt"):
                shutil.copy(os.path.join(train_labels, file), output_folder)

    print("Proses selesai. Semua file sudah digabung di:", output_folder)


# Contoh penggunaan
source = "C:/Users/Aldan/Desktop/ImproTYT/CAMERA 5 INNOVA.v5i.yolov8"  # ganti dengan folder aslimu
output = "C:/Users/Aldan/Desktop/ImproTYT/dataset"    # misalnya: "./dataset_final"
merge_roboflow_dataset(source, output)
```
