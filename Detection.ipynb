```python
import os
def change_extension_to_txt(path):
        if path.lower().endswith('.jpg'):
            # Use os.path.splitext to safely split the path and extension
            # Then rejoin with the new .txt extension
            return os.path.splitext(path)[0] + '.txt'
        return path # Return original path if it's not a .jpg
```

```python
import pandas as pd
import ast
import os
import shutil

# Baca CSV
df = pd.read_csv(r"E:\24_1_Juli_dataset\sorted.csv")

# Folder tujuan utama
destination_folder = r'C:\Users\Aldan\Desktop\ImproTYT\Review03072025'  # ganti sesuai kebutuhan

for index, row in df.iterrows():
    path_gambar = row["Path Gambar"]
    label_str = row["Label YOLO Format"]
    model_name = row["Model"]

    try:
        # Ambil nama file dan parent folder datetime
        nama_file = os.path.basename(path_gambar)  # CAMERA_1.jpg
        nama_camera = os.path.splitext(nama_file)[0]  # CAMERA_1
        parent_folder = os.path.basename(os.path.dirname(path_gambar))  # 2025-07-01_11-09-25

        # Nama baru
        nama_baru = f"{nama_camera}_{parent_folder}"

        # Buat folder model jika belum ada
        model_folder = os.path.join(destination_folder, model_name)
        os.makedirs(model_folder, exist_ok=True)

        # Path tujuan file gambar dan txt
        dst_img_path = os.path.join(model_folder, f"{nama_baru}.jpg")
        dst_txt_path = os.path.join(model_folder, f"{nama_baru}.txt")

        # Copy gambar
        shutil.copy2(path_gambar, dst_img_path)
        print(f"Gambar dicopy ke: {dst_img_path}")

        # Simpan label YOLO format ke txt
        label_list = ast.literal_eval(label_str)
        with open(dst_txt_path, "w") as f:
            for line in label_list:
                f.write(line + "\n")
        print(f"Label disimpan ke: {dst_txt_path}")
        
    except Exception as e:
        print(f"Error di baris {index}: {e}")
```
